```python
#Dataset vgsales.csv

import pandas as pd
```

```python
#leemos los datos del archivo CSV
datos=pd.read_csv('vgsales.csv')
datos
```

```python
#programa que reciba las ventas por rango de años y muestre en pantalla datos de ventas y como se verian con 10% de descuento
#Rango de años
inicio=2000
fin=2004
ventas_rango=datos[(datos['Year']>=inicio) & (datos['Year']<=fin)]
print("Ventas por rango de años:")
print(ventas_rango)
```

```python
#Con el 10% de decuento
ventas_rango['Ventas con descuento']=ventas_rango['Global_Sales']*0.9
#Mostrar las ventas con el descuento
ventas_rango[['Year','Global_Sales','Ventas con descuento']]
```
